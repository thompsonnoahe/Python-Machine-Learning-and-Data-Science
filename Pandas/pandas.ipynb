```python
# Note: function should only return a single value.
def yelp(price):
    if price < 10:
        return '$'
    elif price >= 10 and price <= 30:
        return '$$'
    else:
        return '$$$'
```

```python
df['yelp'] = df['total_bill'].apply(yelp)
```

#### apply() on multiple columns

```python
df.head()
```

```python
def quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return "generous"
    else:
        return "other"
```

```python
df['tip_quality'] = df[['total_bill', 'tip']].apply(lambda df:quality(df['total_bill'], df['tip']), axis=1)
```

```python
df
```

```python
df['tip_quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
```

```python
df
```

## Merging DataFrames

### Inner Join

```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registrations
```

```python
logins
```

```python
help(pd.merge)
```

```python
pd.merge(registrations, logins, how='inner', on='name')
```

### Left, Right, and Outer Joins

#### Note: the order in which the tables are passed in affect the results

```python
pd.merge(left=registrations, right=logins, how='left', on='name')
```

```python
pd.merge(left=registrations, right=logins, how='right', on='name')
```

```python
pd.merge(registrations, logins, how='outer', on='name')
```

```python
registrations = registrations.set_index("name")
```

```python
registrations
```

```python
pd.merge(registrations, logins, left_index=True, right_on='name', how='inner')
```

```python
registrations = registrations.reset_index()
```

```python
registrations
```

```python
registrations.columns = ['reg_name', 'reg_id']
```

```python
registrations
```

```python
results = pd.merge(registrations, logins, how='inner', left_on='reg_name', right_on='name')
```

```python
results
```

```python
results.drop('reg_name', axis=1)
```

```python
registrations.columns = ['name', 'id']
```

```python
logins.columns = ['id', 'name']
```

```python
registrations
```

```python
logins
```

```python
# Pandas will rename the duplicate columns
pd.merge(registrations, logins, how='inner', on='name')
```

```python
pd.merge(registrations, logins, how='inner', on='name', suffixes=('_reg', '_log'))
```
